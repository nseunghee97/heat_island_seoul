## 해발고도

```python
os.chdir('./그리드화')
```

```python
eles = pd.read_csv('그리드_해발고도2.csv')
ele = eles[['fid','ELEV']]
```

```python
# fid당 해발고도 feature 생성
fids = list(ele.fid.unique())
fid_ls, max_ele, min_ele, avg_ele, num_ele, diff_ele = [], [], [], [], [], []
for fid in fids:
    df = ele.loc[ele.fid == fid,:].reset_index(drop=True)
    fid_ls.append(fid)
    max_ele.append(np.max(df.ELEV))
    min_ele.append(np.min(df.ELEV))
    avg_ele.append(np.mean(df.ELEV))
    num_ele.append(len(df.ELEV))
    diff_ele.append(np.max(df.ELEV)-np.min(df.ELEV))
    
ele_df = pd.DataFrame({'fid':fid_ls,'max_ele':max_ele,'min_ele':min_ele,'avg_ele':avg_ele,'num_ele':num_ele,'diff_ele':diff_ele})
ele_df
```

```python
ele_df.to_csv('그리드_해발고도_정리.csv',index=False)
```

```python
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
# 초미세먼지
driver = webdriver.Chrome('C:\\chromedriver.exe')

driver.get('https://www.airkorea.or.kr/web/sidoAirInfo/sidoAirInfoDay30ByLast?itemCode=10008&ymd=2018-04-30&areaCode=02&tDate=2018-04-01&monthDay=30&tYear=2018&tDt=04')
html = driver.page_source
soup = BeautifulSoup(html,'lxml')
cols = soup.select('thead > tr > th')
r_cols = [c.text for c in cols]

dusts = {}
for col in r_cols:
    dusts[col] = []

rows = soup.select('tbody > tr')
for r in rows:
    eles = r.select('tr > td')
    for i in range(len(eles)):
        try:
            put =int(eles[i].text.strip())
        except:
            put = eles[i].text.strip()
        dusts[r_cols[i]].append(put)
 
driver.close()
dust25_df = pd.concat([pd.Series(dusts[col]) for col in list(dusts.keys())],axis=1)
dust25_df.columns = list(dusts.keys())
dust25_df
```

## 가중치 처리

```python
file2_re # 최종본_변수정리_rescale2
file2_re.columns
```

```python
mse = pd.read_csv('MSE 데이터프레임.csv',encoding='cp949')
mse
```

```python
mse['abs_가중치'] = [v/np.max(mse.vars) for v in list(mse.vars)]
MSE = mse.sort_values(by='abs_가중치',ascending=False).reset_index(drop=True)
MSE
```

```python
cols = list(MSE.feature)

weight = []
for col in cols:
    df = file1_re.loc[:,['avg_tmp',col]]
    corr = df.corr().iloc[0,1]
    if corr > 0:
        weight.append(MSE.loc[MSE.feature == col,'abs_가중치'].values[0])
    else:
        weight.append(-1* MSE.loc[MSE.feature == col,'abs_가중치'].values[0])
MSE['weight'] = weight
MSE
```

```python
MSE.loc[MSE.weight > 0,:]
```

```python
MSE.to_csv('그리드_가중치_완성본.csv',index=False)
```

```python
# 클러스터1내에서 열섬척도를 구하고 상위 몇개 뽑기
cluster1 = file1_re.loc[file1_re.label == 0,:].reset_index(drop=True)
cols = list(MSE.feature)

fids = list(cluster1.fid)
fid_ls, heat_p = [], []

for fid in fids:
    fid_ls.append(fid)
    value_sum = 0
    for col in cols:
        value = cluster1.loc[cluster1.fid == fid,col].values[0]
        wei = MSE.loc[MSE.feature == col,'weight'].values [0]
        value_sum += value * wei
    heat_p.append(value_sum)
    
heat_map = pd.DataFrame({'fid':fid_ls,'열섬척도':heat_p})
heat_map
```

```python
heat_map.sort_values(by='열섬척도',ascending=False).reset_index(drop=True)
```
